```python
from ultralytics import YOLO

# Initialize the model (YOLOv8s for example)
model = YOLO('yolov8s.pt')

# Train the model
results = model.train(data='/content/archive/data.yaml', epochs=20, imgsz=640)
```

```python
model_path = 'trained_model.pt'
model.save(model_path)


import cv2
import glob
import os  # Make sure to import the os module
import matplotlib.pyplot as plt


model = YOLO(model_path)

test_images_dir = 'dataset/test/images'

output_dir = 'runs/detect/exp'
os.makedirs(output_dir, exist_ok=True)
```

```python
test_images = glob.glob(os.path.join(test_images_dir, '*.jpg'))

class_names = ['plastic', 'waste', 'oil_spill', 'glass', 'metal']
```

```python
for image_path in test_images:
    image = cv2.imread(image_path)

    results = model(image)

    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = class_names[int(box.cls[0])]
            confidence = box.conf[0]
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(image, f'{label} {confidence:.2f}', (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    output_path = os.path.join(output_dir, os.path.basename(image_path))
    cv2.imwrite(output_path, image)

    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Detections in {os.path.basename(image_path)}")
    plt.axis('off')
    plt.show()
```
